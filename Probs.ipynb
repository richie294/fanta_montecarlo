# Functions

```python
def crea_girone(teams):
    matches = list(itertools.combinations(teams, 2))

    # Create a dictionary to store matchday and games
    matchdays_dict = {}

    # Assign each match to a matchday with 4 games per matchday
    matches_per_matchday = 4
    matchday = 1
    match_idx = 0

    for i in range(1, 8):  # There will be 7 matchdays
        games = []
        for j in range(matches_per_matchday):
            games.append(matches[match_idx])
            match_idx += 1
        matchdays_dict[i] = games

    return matchdays_dict
```

```python
def campionato(teams,num_gironi,s_std, std_lag):
    games_dict = crea_girone(teams)
    classifica = np.array([0]*len(teams))
    
    for j in range(num_gironi):
        classifica_temp = classifica_girone(teams, games_dict, s_std, std_lag)
        classifica += np.array(classifica_temp)
        
    return list(classifica)
```

```python
def main(teams,num_gironi,s_std, std_lag, num_sim):
    # create dictionary for counting number of championship wins
    count_dict = {}
    for j in range(len(teams)):
        count_dict[j] = 0
        
    # simulate championship
    for k in range(num_sim):
        print(f'Running Sim {k} of {num_sim}')
        classifica = campionato(teams,num_gironi,s_std,std_lag)
        winner_index = classifica.index(max(classifica))
        count_dict[winner_index] += 1
    
    count_list = list(count_dict.values())
    probs = np.array(count_list)/num_sim
    
    return probs
```
